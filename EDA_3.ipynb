```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

pd.options.display.max_columns = None
pd.options.display.max_colwidth = None

data = pd.read_csv('data/data_normalized.csv')
```

```python
data.head()
```

```python
data_pivot = data.melt(id_vars=['countrycode', 'country', 'year'])
```

```python
data_pivot['variable'] = data_pivot['variable'] + '_' + data_pivot['year'].astype(str)
data_pivot.drop('year', axis=1, inplace=True)
data_pivot.head()
```

```python
data_pivot = data_pivot.groupby(['countrycode', 'country', 'variable'])['value'].aggregate('mean').unstack()
```

```python
data_pivot.head()
```

```python
data_pivot.to_csv('data/data_flat.csv')
```
